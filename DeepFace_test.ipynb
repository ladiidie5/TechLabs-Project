```
from deepface import DeepFace
import cv2
import matplotlib.pyplot as ptl
```

```
img_path = 'img7.jpg'
ptl.imshow(cv2.imread(img_path))
```

```
em1 = DeepFace.analyze(img_path)
```

```
print ('Emotion detected ' + em1['dominant_emotion'])
```
